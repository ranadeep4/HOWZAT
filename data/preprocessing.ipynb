```python
import wikipedia
```

```python
def get_wiki_summary(name, sentences=3):
    try:
        # search for the player
        results = wikipedia.search(name)
        if not results:
            return ""
        
        # choose first result (most common)
        page_title = results[0]
        
        # fetch summary
        summary = wikipedia.summary(page_title, sentences=sentences)
        return summary
    
    except wikipedia.exceptions.DisambiguationError:
        return ""
    except wikipedia.exceptions.PageError:
        return ""
    except Exception:
        return ""
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.head()
```

```python
from tqdm import tqdm
```

```python
summaries = []

for name in tqdm(df["Player"]):
    summaries.append(get_wiki_summary(name))

df["summary"] = summaries
df.head()
```

```python
df.to_csv("ipl_with_summary.csv", index=False)
```
